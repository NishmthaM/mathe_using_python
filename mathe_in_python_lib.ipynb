Scalar,Vector,Matrix and Tensor in numpy

```python
#scalar- single number
scalar = 5.0
print(scalar)
print("Type of scalar:", type(scalar))
```

```python
#vector - 1D data

vector = np.array([5,8,6])
print("Type of vector:",type(vector))
print("vector:",vector)
print("Shape of the vector:",vector.shape)
```

```python
#matrix -2D data

matrix = np.array(
    [
      [2,6,4],
      [1,3,5]
    ])
print("Type of matrix:",type(matrix))
print("matrix:",matrix)
print("Shape of the matrix:",matrix.shape)
```

```python
#Tensor - more than 2D data

tensor = np.array(
    [[[
      [2,6,4],
      [1,3,5]
    ],
    [
        [2.6, 3.6, 2.8],
        [1.1, 3.1, 2.2]
    ],

    [
        [1, 6, 5],
        [6, 8, 6]
    ]
      ]])
print("Type of tensor:",type(tensor))
print("tensor:",tensor)
print("Shape of the tensor:",tensor.shape)
print(tensor.ndim)
```

Basic operations: addition, scalar multiplication,transpose


```python
import numpy as np

a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.array ([[2,6,4],
      [1,3,5]])
print("a:", a)
print("b:", b)

print("")

add= a+b
print("vector addition :",add)

print("")

# scaling up
scalar= 2*a
print("Scalar multiplication 2*a :",scalar)

# scaling down

scalar1=0.5 * a
print("Scalar multiplication 0.5*a",scalar1)

print("")

#transpose

trans=c.T
print("Transpose of matrix",trans)
print("Shape of transpose matrix:",trans.shape,"| Shape of matrix",c.shape)
```

Dot product


```python
v1=np.array([2,3,6])
v2=np.array([4,2,7])

print("V1:",v1)
print("V2:",v2)

```

```python
dot=np.dot(v1,v2)
print("Dot product v1.v2:",dot)
```

Euclidean distance

```python
diff= v1-v2
print("Diffrence v1-v2:",diff)
```

```python
distance=np.linalg.norm(diff)
print("Euclidians distance :",distance)
```

Matrix multiplication linear model

```python
X= np.array([
    [0.2, 0.3],
    [0.6, 0.4],
    [0.5, 0.7]
])

print("X:\n",X)
print("X shape:", X.shape)

```

```python
w=np.array([
    [0.5],
    [0.6]
])
print("w:\n",w)
print("w shape:", w.shape)
```

```python
Xw= X @ w # np.dot
print("Xw:\n",Xw)
print("Xw shape:", Xw.shape)
```

Sample dataset


```python
data=np.array([
    [85, 26.6, 31, 0],
    [183, 23.3, 32, 1],
    [89, 28.1, 21, 0],
    [137, 43.1, 33, 1]
])

print("Data matrix:\n",data)
print("Data shape:",data.shape)

```

```python
x_features= data[:,: -1]   # : → take all rows # :-1 → take all columns except last

y_target= data[:, -1]  # -1 represents give only last column

print("Features\n:",x_features)
print()
print("Features shape:",x_features.shape)
print()

print("Taget vector:",y_target)
print()
print("Target shape",y_target.shape)
```
